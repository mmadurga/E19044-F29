# N=20 ESPES 

Let's look at the effective single particle states and gaps for N=20 

```julia
using DelimitedFiles, Plots
```

## SDPF-MU

First, single particle energies with modern SDPF-MU

```julia


sdpfmuespesN20 = readdlm("n20ESPE.txt",comments=true,comment_char='#')
sdpfmuespesN16 = readdlm("n16ESPE.txt",comments=true,comment_char='#')
```

### N=20

let's start with effective single particle energies

```julia
plot(sdpfmuespesN20[:,1],sdpfmuespesN20[:,12],label="s1/2",lw=3,linecolor=:black,xlabel="Z",ylabel="E (MeV)")
plot!(sdpfmuespesN20[:,1],sdpfmuespesN20[:,11],label="d3/2",lw=3,linecolor=:red)
plot!(sdpfmuespesN20[:,1],sdpfmuespesN20[:,13],label="f7/2",lw=3,linecolor=:blue)
annotate!(11.5,2,"N=20 isotopes")
```

What about the shell gaps

```julia
plot(sdpfmuespesN20[:,1],sdpfmuespesN20[:,11].-sdpfmuespesN20[:,12],lw=3,linecolor=:black,linestyle=:dash,xlabel="Z",ylabel="ΔE (MeV)",label="N=16 gap")
plot!(sdpfmuespesN20[:,1],sdpfmuespesN20[:,13].-sdpfmuespesN20[:,11],lw=3,linecolor=:red,linestyle=:dash,label="N=20 gap")
annotate!(10.5,6.5,"N=20 isotopes")
```

### N=16 isotopes

same excercise for N=16 isotopes

```julia
plot(sdpfmuespesN16[:,1],sdpfmuespesN16[:,12],label="s1/2",lw=3,linecolor=:black,xlabel="Z",ylabel="E (MeV)")
plot!(sdpfmuespesN16[:,1],sdpfmuespesN16[:,11],label="d3/2",lw=3,linecolor=:red)
plot!(sdpfmuespesN16[:,1],sdpfmuespesN16[:,13],label="f7/2",lw=3,linecolor=:blue)
annotate!(11.5,3,"N=16 isotopes")
```

```julia
plot(sdpfmuespesN16[:,1],sdpfmuespesN16[:,11].-sdpfmuespesN16[:,12],lw=3,linecolor=:black,linestyle=:dash,xlabel="Z",ylabel="ΔE (MeV)",label="N=16 gap")
plot!(sdpfmuespesN16[:,1],sdpfmuespesN16[:,13].-sdpfmuespesN16[:,11],lw=3,linecolor=:red,linestyle=:dash,label="N=20 gap")
annotate!(10.5,6.7,"N=16 isotopes")
```

### SDPF-MU gaps 

Gaps for both N=16 and N=20

```julia
plot(sdpfmuespesN16[:,1],sdpfmuespesN16[:,11].-sdpfmuespesN16[:,12],lw=3,linecolor=:black,xlabel="Z",ylabel="ΔE (MeV)",label="N=16 gap N=16 isotopes")
plot!(sdpfmuespesN16[:,1],sdpfmuespesN16[:,13].-sdpfmuespesN16[:,11],lw=3,linecolor=:red,label="N=20 gap N=16 isotopes")
plot!(sdpfmuespesN20[:,1],sdpfmuespesN20[:,11].-sdpfmuespesN20[:,12],lw=3,linecolor=:black,linestyle=:dash,xlabel="Z",ylabel="ΔE (MeV)",label="N=16 gap N=20 isotopes")
plot!(sdpfmuespesN20[:,1],sdpfmuespesN20[:,13].-sdpfmuespesN20[:,11],lw=3,linecolor=:red,linestyle=:dash,label="N=20 gap N=20 isotopes")
```
