## Study of modifications to USDB to match the experimental proton knockout spectroscopic factor

Experiments at [RIKEN](https://doi.org/10.1103/PhysRevLett.124.212502) and [NSCL](https://doi.org/10.1103/PhysRevC.106.L061303) report the spectroscopic factor of a single proton knockout ($C^2S$) from $^{25}F$ to the ground state of $^{24}O$ at around 0.3. 

I calculated the spectroscopic factor using USDB and some modifications of the proton neutron interaction to reproduce this effect.

**Summary:** Any modification of the strength large enough to reproduce the $C^2S$ reduction observed in $^{25}F$ 'breaks' the surrounding isotopes or isotones. There is no consistent way to reproduce the experimental results, leaving the changes needed to the single particle orbitals in $^{25}F$ alone in the entire region. At the moment it is hard to envision a change in the nuclear structure that can act on $^{25}F$ without affecting the neighboring nuclei.

### Regular USDB

Usinf the regular USDB I get a $C^2S$ of 0.9 to the ground state of $^{24}O$, with the rest fragmented to the second and third $0^+$ states in oxygen.

### USDB with increased $\pi d_{5/2}-\nu d_{3/2}$ interaction

Followig the prescription in [here](https://doi.org/10.1103/PhysRevLett.124.212502), I increased the interaction between the $\pi d_{5/2}-\nu d_{3/2}$ levels in order to lower the $\nu d_{3/2}$ state by 3 MeV. 

The $C^2S$ to the ground state of $^{24}O$ is now 0.37, with 0.4 going to the second $0^+$ state. This matches the experimental result at RIKEN were a substantial amount of strength populates unbound states in $^{24}O$, and reflects the NSCL result where the ground state of $^{25}F$ has a substantial component from polarized $^{24}O$

### USDB with consistent tensor force change between the $\pi d$ and $\nu d$ orbitals

Following Otsuka's observation, one needs to decrease the interaction strength between the $\pi d_{5/2}$ and $\nu d_{5/2}$ if one wants to increase the $\pi d_{5/2}$ and $\nu d_{3/2}$. However doing this pushes the proton $s_{1/2}$ to the fermi surface of Z=9 isotopes, against the experimental evidence. With the small change showed below the ground state of $^{25}F$ is still a 5/2+ state, however the $C^2S$ is still 0.77, compared to the ~0.4 experimetal one. In order to make the $C^2S$ between the $5/2^+$ state in $^{25}F$ as small as the experimental one that state becomes an excited state, with the ground state being a $1/2^+$.
